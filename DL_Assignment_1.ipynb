## Question_5

```python
import matplotlib.pyplot as plt
import numpy as np
import math
  
```

```python
x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))
  
plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
  
plt.show()
```

```python
in_array = np.linspace(-10, 10, 22)
out_array = np.tanh(in_array)

 
print("in_array : ", in_array)
print("\nout_array : ", out_array)
  
# red for numpy.tanh()
plt.plot(in_array, out_array, color = 'blue', marker = "o")
plt.title("numpy.tanh()")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python

def relu(x):
    return(np.maximum(0, x))

x_values = np.linspace(start = -10, stop = 10, num = 22)
print(x_values)
relu_values = relu(x_values)

x = x_values
y = relu_values
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("RELU(X)")
plt.show()
```
